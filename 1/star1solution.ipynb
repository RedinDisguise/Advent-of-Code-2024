Access the file and numbers

```python
input = open("input.txt", "r")
list1 = []
list2 = []

for x in input:
  list1.append(int(x[:5]))
  list2.append(int(x[8:13]))

```

Sort into two lists of ascending value

```python
list1.sort(reverse = False)
list2.sort(reverse = False)
```

Calculate length of each

```python
length=0
index=0

print(list1)
print(list2)

for x in list1:
    length = length + abs(x - list2[index])
    
print(length)
```
